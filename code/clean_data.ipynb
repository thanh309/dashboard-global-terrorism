```python
def _decode(o):
    if isinstance(o, str):
        try:
            return int(o)
        except ValueError:
            return o
    elif isinstance(o, dict):
        return {_decode(k): v for k, v in o.items()}
    else:
        return o

import json
with open('utils/map_country_to_iso.json', 'r') as file:
    mapping = json.load(file, object_hook=_decode)

data['country'] = data['country'].map(mapping)
```

```python
data['full_datetime'] = data.apply(lambda row: f"{row['iyear']}-{row['imonth']:02d}-{row['iday']:02d}", axis=1)
```
