```python
data.rename(columns={'iyear':'year',                    'extended':'is_extended',               'alternative_txt':'attack_tactic',      'attacktype1_txt':'attack_type',
                     'targtype1_txt':'target_type',     'targsubtype1_txt':'target_subcat',     'gname':'te_group',                     'motive':'te_motive',
                     'weaptype1_txt':'weapon_type',     'weapsubtype1_txt':'weapon_subtype',    'nkill':'total_killed',                 'nwound':'total_wounded',
                     'nkillter':'te_killed',            'nwoundte':'te_wounded',                'propextent_txt':'prorperty_damage',    'hostage_con':'ishostkid',
                     'ransomamt':'ransom_amt',          'ransompaid':'ransom_amt_paid',         'hostkidoutcome_txt':'hostage_outcome'
                    }, inplace=True)
```

```python
cleaned_data = data[['year',            'is_extended',      'city',         'country_txt',      'region_txt',       'summary',          'success',      'attack_tactic',
                     'attack_type',     'suicide',          'target_type',  'target_subcat',    'te_group',         'te_motive',        'weapon_type',  'weapon_subtype',
                     'total_killed',    'total_wounded',    'te_killed',    'te_wounded',       'prorperty_damage', 'ishostkid',        'ransom',       'ransom_amt',
                     'ransom_amt_paid', 'hostage_outcome']].reset_index(drop=True)
```

```python
cleaned_data['city'].fillna('Unknown', inplace=True)

cleaned_data['summary'].fillna('Unknown', inplace=True)

cleaned_data['attack_tactic'].fillna('Unknown', inplace=True)

cleaned_data['target_subcat'].fillna('Unknown', inplace=True)

cleaned_data['te_motive'].fillna('Unknown', inplace=True)

cleaned_data['weapon_subtype'].fillna('Unknown', inplace=True)



cleaned_data['total_killed'] = cleaned_data['total_killed'].fillna(cleaned_data['total_killed'].mode()[0])

cleaned_data['total_wounded'] = cleaned_data['total_wounded'].fillna(cleaned_data['total_wounded'].mode()[0])

cleaned_data['te_killed'] = cleaned_data['te_killed'].fillna(cleaned_data['te_killed'].mode()[0])

cleaned_data['te_wounded'] = cleaned_data['te_wounded'].fillna(cleaned_data['te_wounded'].mode()[0])


cleaned_data = cleaned_data.astype({'total_killed':'int', 'total_wounded':'int', 'te_killed':'int', 'te_wounded':'int'})


cleaned_data['civ_killed'] = cleaned_data['total_killed'] - cleaned_data['te_killed']

cleaned_data['civ_wounded'] = cleaned_data['total_wounded'] - cleaned_data['te_wounded']

cleaned_data['total_casualities'] = cleaned_data['total_killed'] + cleaned_data['total_wounded']



cleaned_data['prorperty_damage'].fillna('Unknown', inplace=True)

cleaned_data['ishostkid'].fillna(-9.0, inplace=True)
cleaned_data['ishostkid'] = cleaned_data['ishostkid'].map(lambda x: -9.0 if x < 0 else x)

cleaned_data['ransom'].fillna(-9.0, inplace=True)
cleaned_data['ransom'] = cleaned_data['ransom'].map(lambda x: -9.0 if x < 0 else x)

cleaned_data['ransom_amt'].fillna(0, inplace = True)
cleaned_data['ransom_amt'] = cleaned_data['ransom_amt'].map(lambda x: -9.0 if x < 0 else x)

cleaned_data['ransom_amt_paid'].fillna(0, inplace = True)
cleaned_data['ransom_amt_paid'] = cleaned_data['ransom_amt_paid'].map(lambda x: -9.0 if x < 0 else x)

cleaned_data = cleaned_data.astype({'ransom_amt':'int', 'ransom_amt_paid':'int', 'ishostkid':'int', 'ransom':'int'})

cleaned_data['hostage_outcome'].fillna('Unknown', inplace=True)

cleaned_data = cleaned_data.astype({'year':'int', 'is_extended':'int', 'success':'int', 'suicide':'int'})
```

```python
cleaned_data['weapon_type'] = cleaned_data['weapon_type'].replace({'Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)':'Vehicle'})

cleaned_data['attack_type'] = cleaned_data['attack_type'].replace({'Hostage Taking (Kidnapping)':'Kidnapping','Hostage Taking (Barricade Incident)':'Barricade Incident'})
```

```python
cleaned_data = clean_country(cleaned_data, 'country_txt', output_format='alpha-3')

cleaned_data.rename(columns = {'country_txt_clean':'country_code'}, inplace = True)
```

```python
cleaned_data = cleaned_data.dropna()
```
